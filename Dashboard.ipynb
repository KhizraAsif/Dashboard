```python
import pandas as pd
import plotly.express as px
import streamlit as st
from streamlit_jupyter import StreamlitPatcher, tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
# Patch Streamlit so it works inside Jupyter
StreamlitPatcher().jupyter()
```

```python
#import dataset
df = pd.read_csv("D:/New folder/superstore.csv")
```

```python
#Shape of the dataset (rows, columns)
st.write(df.shape)
```

```python
#Columns names
st.write("\nColumns:\n", df.columns.tolist())
```

```python
#First Five Rows
st.write(df.head())
```

```python
# Check missing values
st.write(df.isnull().sum())
```

```python
st.write("Data Types of Each Column:")
st.write(df.dtypes)
```

```python
# Convert datatypes
df["Order.Date"] = pd.to_datetime(df["Order.Date"])
df["Ship.Date"] = pd.to_datetime(df["Ship.Date"])
df["Sales"] = df["Sales"].astype(float)
```

```python
# Show confirmation in Streamlit
st.write("Data Types After Conversion:")
st.write(df.dtypes)
```

```python
st.sidebar.header("Filters")

region = st.sidebar.multiselect(
    "Select Region:",
    options=df["Region"].unique(),
    default=df["Region"].unique()
)

category = st.sidebar.multiselect(
    "Select Category:",
    options=df["Category"].unique(),
    default=df["Category"].unique()
)

sub_category = st.sidebar.multiselect(
    "Select Sub-Category:",
    options=df["Sub.Category"].unique(),  
    default=df["Sub.Category"].unique()    
)

# Apply filters
filtered_df = df[
    (df["Region"].isin(region)) &
    (df["Category"].isin(category)) &
    (df["Sub.Category"].isin(sub_category))  
]
```

```python
total_sales = filtered_df["Sales"].sum()
total_profit = filtered_df["Profit"].sum()

st.title(" Global Superstore Dashboard")

col1, col2 = st.columns(2)
col1.metric("Total Sales", f"${total_sales:,.0f}")
col2.metric("Total Profit", f"${total_profit:,.0f}")
```

```python
# Sales by Category
sales_by_category = (
    filtered_df.groupby("Category")["Sales"].sum().reset_index()
)
fig_category = px.bar(sales_by_category, x="Category", y="Sales", title="Sales by Category")
st.plotly_chart(fig_category, use_container_width=True)

# Profit by Sub-Category
profit_by_subcat = (
    filtered_df.groupby("Sub.Category")["Profit"].sum().reset_index()
)
fig_subcat = px.bar(profit_by_subcat, x="Sub.Category", y="Profit", title="Profit by Sub-Category")
st.plotly_chart(fig_subcat, use_container_width=True)

# Sales Trend
sales_trend = (
    filtered_df.groupby("Order.Date")["Sales"].sum().reset_index()
)
fig_trend = px.line(sales_trend, x="Order.Date", y="Sales", title="Sales Trend Over Time")
st.plotly_chart(fig_trend, use_container_width=True)
```

```python
top_customers = (
    filtered_df.groupby("Customer.Name")["Sales"].sum()
    .nlargest(5).reset_index()
)

st.subheader(" Top 5 Customers by Sales")
st.dataframe(top_customers)
```
